def convert_to_12_hour(time_str):
    try:
        # Convert to datetime to check the format
        time_obj = pd.to_datetime(time_str, format='%H:%M:%S', errors='raise')
        # Check if it's in 24-hour format and convert to 12-hour format
        return time_obj.strftime('%I:%M %p') if time_obj.hour < 12 else time_obj.strftime('%I:%M %p')
    except ValueError:
        # If not in 24-hour format, return the original string
        return time_str

# Apply the conversion function to the 'Time of Restoration' column
df['Time of Restoration'] = df['Time of Restoration'].apply(convert_to_12_hour)
